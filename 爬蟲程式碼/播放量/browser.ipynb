```python
from cgitb import text
from curses import KEY_ENTER
from pickletools import uint1
from xml.dom.minidom import Element
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
```

```python
browser=webdriver.Chrome()
browser.get('https://www.bilibili.com/v/popular/all?spm_id_from=333.1007.0.0')
try:
    wait=WebDriverWait(browser,100)
    element=wait.until(EC.presence_of_all_elements_located)
finally:
    print('加载完成')

windows=browser.window_handles 
browser.switch_to.window(windows[-1])
browser.execute_script('var q=document.documentElement.scrollTop=0')
button_1=browser.find_element_by_xpath('//*[@id="app"]/div/div[1]/div[2]/div/span') # 每周必看
button_1.click()
time.sleep(2)
button_2=browser.find_element_by_xpath('//*[@id="app"]/div/div[2]/div[1]/div[1]/span')
button_2.click()
time.sleep(2)
button_3=browser.find_element_by_xpath('//*[@id="app"]/div/div[2]/div[1]/div[1]/div/div[4]/p[1]')
button_3.click()

windows=browser.window_handles
browser.switch_to.window(windows[-1])
try:
    wait=WebDriverWait(browser,100)
    element=wait.until(EC.presence_of_all_elements_located)
finally:
    print('加载完成')

time.sleep(2)

pageviews = []
title_week = [] # 每周必看视频的标题

for i in range (1,43):
    try: 
        windows=browser.window_handles
        browser.switch_to.window(windows[-1])
        elem_text=browser.find_element_by_xpath(f'//*[@id="app"]/div/div[2]/div[2]/div/div[{i}]/div[2]/p')
        elem_browser=browser.find_element_by_xpath(f'//*[@id="app"]/div/div[2]/div[2]/div/div[{i}]/div[2]/div/p/span[1]')
        print(elem_text.text)
        print(elem_browser.text)
        title_week.append(elem_text.text)
        pageviews.append(elem_browser.text)
        
    except Exception as e:
        pass
    continue
```

```python
for i in range(0,42):
    pageviews[i] = pageviews[i].strip('万')
    print(pageviews[i])
```

```python
dict_week = dict(zip(title_week,pageviews))

# label_csv = ['视频标题','播放量（万）']
with open('play-volume.csv', 'w',encoding='utf_8_sig') as f:  
    writer = csv.writer(f)
    for k, v in dict_week.items():
       writer.writerow([k, v])
```

```python
from wordcloud import WordCloud
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('play-volume.csv',names=['标题','播放量'])

title_week = data['标题']
pageviews = data['播放量']

dict_week = dict(zip(title_week,pageviews))
dict_week
```

```python
import matplotlib.colors as colors  # 处理图片相关内容

colormaps = colors.ListedColormap(['#FF7F50','#FF0000','#FFE4C4'])

my_cloud = WordCloud(
    colormap=colormaps,  # 指定颜色
    background_color='white',  # 设置背景颜色  默认是black
    width=1800, height=1200,
    max_words=100,            # 词云显示的最大词语数量
    font_path='Arial Unicode.ttf',   # 设置字体  显示中文
    max_font_size=99,         # 设置字体最大值
    min_font_size=16,         # 设置子图最小值
    random_state=50           # 设置随机生成状态，即多少种配色方案
).generate_from_frequencies(dict_week)

plt.subplots(figsize=(10,6),dpi=300)
plt.imshow(my_cloud, interpolation='bilinear')
# 显示设置词云图中无坐标轴
plt.axis('off')
plt.show()
```
