```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
import time
import re

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=chrome_options)
```

```python
driver.get("https://www.bilibili.com/v/popular/all")

WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

title_all = [] # 综合热榜视频的标题
label_all = [] # 综合热榜视频的标签

for k in range (1,201):
    label_all.append([]) # 生成子列表 用于存储每个视频的标签

    windows=driver.window_handles # 获取当前窗口句柄集合（列表类型）
    driver.switch_to.window(windows[0]) #进入第一个主要窗口

    title_name = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div[2]/div/ul/div[' + str(k) + ']/div[2]/p')  # 标题名称 
    title_all.append(title_name.text) #把当前标题存入列表中

    title_name.click() # 点击视频

    windows=driver.window_handles
    driver.switch_to.window(windows[-1]) #进入最右边弹出窗口
    WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

    tags = driver.find_elements(by=By.CLASS_NAME, value="tag") # 标签内容

    for tag in tags:
      label_all[k-1].append(tag.text) #把当前标签存入列表中
    driver.close()
    
driver.quit()
```

```python
driver = webdriver.Chrome('chromedriver',options=chrome_options)
driver.get("https://www.bilibili.com/v/popular/weekly?num=164")

WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

title_weekly = [] # 每周必看视频的标题
label_weekly = [] # 每周必看视频的标签

for k in range (1,43):

    windows=driver.window_handles # 获取当前窗口句柄集合（列表类型）
    driver.switch_to.window(windows[0]) #进入第一个主要窗口

    title_name = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div[2]/div[2]/div/div[' + str(k) + ']/div[2]/p')  # 标题名称 
    title_weekly.append(title_name.text) #把当前标题存入列表中
    title_name.click() # 点击视频

    windows=driver.window_handles
    driver.switch_to.window(windows[-1]) #进入最右边弹出窗口
    WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

    tags = driver.find_elements(by=By.CLASS_NAME, value="tag-area") # 标签内容
    for tag in tags:
      label_weekly.append(tag.text.split("\n")) #把当前标签存入列表中
    driver.close()

driver.quit()
```

```python
driver.get("https://www.bilibili.com/v/popular/rank/all")

WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

title_rank = [] # 排行榜视频的标题
label_rank = [] # 排行榜视频的标签

for k in range (1,101):
    label_rank.append([]) # 生成子列表 用于存储每个视频的标签

    windows=driver.window_handles # 获取当前窗口句柄集合（列表类型）
    driver.switch_to.window(windows[0]) #进入第一个主要窗口

    title_name = driver.find_element(by=By.XPATH, value='//*[@id="app"]/div/div[2]/div[2]/ul/li[' + str(k) + ']/div/div[2]/a')  # 标题名称 
    title_rank.append(title_name.text) #把当前标题存入列表中

    title_name.click() # 点击视频

    windows=driver.window_handles
    driver.switch_to.window(windows[-1]) #进入最右边弹出窗口
    WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

    tags = driver.find_elements(by=By.CLASS_NAME, value="tag") # 标签内容

    for tag in tags:
      label_rank[k-1].append(tag.text) #把当前标签存入列表中
    driver.close()
    
driver.quit()
```

```python
#标签词频分析
def freq(label_list):
  label_freq = {}  #建立空字典
  for num in range(len(label_list)):
    for item in label_list[num]:
      if item in label_freq:
        label_freq[item] = label_freq[item] + 1 #若字典内已经有该标签，则字词频率增加1
      else:
        label_freq[item] = 1  #反之，则在字典内建立该标签的值为1
  return label_freq  #返回词频字典

label_all_freq = freq(label_all)
label_weekly_freq = freq(label_weekly)
label_rank_freq = freq(label_rank)

a = sorted(label_all_freq.items(), key=lambda x:-x[1])[:10]   #依照词频高地排序，并取前10个
w = sorted(label_weekly_freq.items(), key=lambda x:-x[1])[:10]
r = sorted(label_rank_freq.items(), key=lambda x:-x[1])[:10]
print("综合视频词频前10名:", a, sep="\n")
print("")
print("每周必看词频前10名:", w, sep="\n")
print("") 
print("排行榜词频前10名:", r, sep="\n")
```

```python
#每周必看视频弹幕截取

def get_cid(url):  #拿取网页cid
  driver.get(url)
  WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)
  html = driver.page_source
  cid = re.findall('https://upos-hz-mirrorakam.akamaized.net/upgcxcode/\d{2}/\d{2}/\d{9}',html)  #透过re找到包含cid的html
  cid_num = cid.pop()  #将所有找到的放在列表并取出最后一个
  cid_num = cid_num[-9:]  #保留最后9位cid

  return cid_num

def get_danmu(cid):
  danmu = []
  url = "https://comment.bilibili.com/" + str(cid) + ".xml"  #输入视频cid能取得所有弹幕的url

  request = requests.get(url)
  request.encoding='utf8'
  bs = BeautifulSoup(request.text, 'html.parser')
  results = bs.find_all('d')
  for result in results:
    danmu.append(result.text)
  return danmu

driver.get("https://www.bilibili.com/v/popular/weekly?num=164")
WebDriverWait(driver,100).until(EC.presence_of_element_located((By.CLASS_NAME,'no-more')))

windows=driver.window_handles # 获取当前窗口句柄集合（列表类型）
driver.switch_to.window(windows[0]) #进入第一个主要窗口

urlList = []  #创建url列表
danmu_weekly = []  #创建弹幕列表
for k in range(1,43):
  names = driver.find_elements(by=By.XPATH, value='//*[@id="app"]/div/div[2]/div[2]/div/div[' + str(k) + ']/div[1]/a') # url名称

  for name in names:
    url = name.get_attribute("href")
    urlList.append(url)

for i in range(len(urlList)):
  url = urlList[i]
  cid = get_cid(url)
  danmu = get_danmu(cid)
  danmu_weekly.append(danmu)
```

```python
#综合热门视频弹幕截取
driver.get("https://www.bilibili.com/v/popular/all")
WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)

for j in range(10):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
  time.sleep(2)


urlList = []
danmu_all = []
for k in range(1,201): 
  names = driver.find_elements(by=By.XPATH, value='//*[@id="app"]/div/div[2]/div/ul/div[' + str(k) + ']/div[1]/a')
  for name in names:
    url = name.get_attribute("href")
    urlList.append(url)


L = []
cidL = []
for i in range(len(urlList)):  #不知道为啥用上面的get_cid函数会报错，所以重新写一次
  url = urlList[i]
  driver.get(url)
  WebDriverWait(driver,100).until(EC.presence_of_all_elements_located)
  html = driver.page_source
  cid = re.findall('https://upos-hz-mirrorakam.akamaized.net/upgcxcode/\d{2}/\d{2}/\d{9}',html)
  for c in cid:
    if len(L) == i+1:
      continue
    else:
      L.append(c)
      cn = re.findall('https://upos-hz-mirrorakam.akamaized.net/upgcxcode/\d{2}/\d{2}/\d{9}',c)
      cid = cn.pop()
      cid = cid[-9:]
      cidL.append(cid)
for cid in cidL:
  danmu = get_danmu(cid)
  danmu_all.append(danmu)
```
